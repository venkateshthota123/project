```python
import pandas as pd
import string
import random

def generate_random_name():
    first_names = ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hank", "Ivy", "Jack"]
    last_names = ["Smith", "Johnson", "Williams", "Jones", "Brown", "Davis", "Miller", "Wilson", "Moore", "Taylor"]
    return f"{random.choice(first_names)} {random.choice(last_names)}"

# Generate random data for 100 students
data = {
    'student_name': [generate_random_name() for _ in range(100)],
    'student_marks': [random.randint(50, 100) for _ in range(100)],
    'student_id': ["".join(random.choices(string.ascii_uppercase + string.digits, k=8)) for _ in range(100)]
}
```

```python
data_frame=pd.DataFrame(data)
data_frame.head()
```

print(data_frame[data_frame['student_marks']>70])

```python
print(data_frame['student_marks'].mean())
print(data_frame['student_marks'].min())
print(data_frame['student_marks'].max())
print(data_frame['student_marks'].median())
print(data_frame['student_marks'].mode())
```

```python
df=data_frame[data_frame['student_marks']>50].sort_values(by=['student_marks','student_name'])
df
```

```python
def highest_marks(student_marks):
  if student_marks>=75:
    return 'A'
  elif student_marks>=65:
    return 'B'
  elif student_marks>=45:
    return 'C'
  elif student_marks>=30:
    return 'D'
  else:
    return 'F'
```

```python
data_frame['Grades']=data_frame['student_marks'].map(highest_marks)
data_frame
```
